#### 1) Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?

```python
nrows = 1625
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close[900:1624]
#ys = df.Close.values
wave = thinkdsp.Wave(ys, framerate=1)
pink_noise = thinkdsp.PinkNoise().make_wave(duration = nrows, framerate=1)
b_noise = thinkdsp.BrownianNoise().make_wave(duration = nrows, framerate=1)
wave.plot()
```

```python
spec = b_noise.make_spectrum()
integ = spec.make_integrated_spectrum()
spec.plot_power(low=1, linewidth=1, alpha=.5)
thinkplot.show(xlabel='freq',ylabel='power density',xscale = 'log',yscale='log',title='brownian')
```

```python
spec = pink_noise.make_spectrum()
integ = spec.make_integrated_spectrum()
spec.plot_power(low=1, linewidth=1, alpha=.5)
thinkplot.show(xlabel='freq',ylabel='power density',xscale = 'log',yscale='log',title='pink')
```

```python
spec = wave.make_spectrum()
integ = spec.make_integrated_spectrum()
spec.plot_power(low=1, linewidth=1, alpha=.5)
thinkplot.show(xlabel='freq',ylabel='power density',xscale = 'log',yscale='log',title='bitcoin')
```

I think that the power spectrum of the bitcoin price wave resembles Brownian noise more because of the steepness of the slope as the power density drops across the frequencies. The pink noise is flatter than the other two.

#### 2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
lags, corrs = autocorr.autocorr(wave)
thinkplot.plot(lags,corrs)
thinkplot.show(xlabel='lag',ylabel='correlation')
```

There are what seem to be 2 "exponential" dips in the curves. However, because the correlation of the signal never returns to one or repeats in any manner, it's safe to say that given the data we have, there is no periodic behavior in the BitCoin prices.

#### 3) Use autocorrelation to estimate the pitch of the saxophone recording in 100475__iluppai__saxophone-weep.wav.  What happens if you filter out the fundamental tone?  Is the estimate based on the autocorrelation function consistent with your perceived pitch?

```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav').segment(6,2)
waves = wave.segment(0,.02)
lags, corrs = autocorr.autocorr(waves)
thinkplot.plot(lags,corrs)
thinkplot.show(xlabel='lag',ylabel='correlation')
wave.make_audio()
```

In order to figure out the pitch, I notice that the correlation seems to return to 1 at about 80 lag, and continues to do so periodically every 80 lag. Knowing that wave.framerate = 44100, we can divide the lag by the framerate, this will give us the period. Inverting the period gives us the frequency or pitch which is approximately 551 hz, and this is consistent with the fundamental frequency in the spectrum below.

```python
spec = wave.make_spectrum()
spec.plot()
thinkplot.config(xlim=[0,5000])
thinkplot.show(xlabel='freq',ylabel='amp')
```

Let's filter out that fundamental, and see what happens to the correlation function and the sound file.

```python
news = spec
news.band_stop(0,900)
new = spec.make_wave()
waves = new.segment(0,.02)
lags, corrs = autocorr.autocorr(waves)
thinkplot.plot(lags,corrs)
thinkplot.show(xlabel='lag',ylabel='correlation')
new.make_audio()
```

Remarkable! The correlation is still periodic according to the 80 lag, and the percieved pitch still lingers though we have taken it out. This provides some merit for the claim that autocorrelation is a possible explanation for the perceived pitch phenomenon.

#### 4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    #result = %timeit -o scipy.fftpack.dct(ys)
    results.append(result)
bests = [result.best for result in results]
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
t
```

For analyze1, the correlation with the log function is strong, meaning that this has a logarythmic complexity associated with it.

#### 5) Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
wave = thinkdsp.read_wave('Alan_Allan_Allen_Steve_WHAAAAT_Best_ending_.wav')
thresh = 10000
seg1 = wave.segment(0,1)
seg2 = wave.segment(1,1)
seg3 = wave.segment(2,1)
seg4 = wave.segment(3,1)
d1 = [0 if x>=thresh else x for x in scipy.fftpack.dct(seg1.ys.astype(float))]
d2 = [0 if x>=thresh else x for x in scipy.fftpack.dct(seg2.ys.astype(float))]
d3 = [0 if x>=thresh else x for x in scipy.fftpack.dct(seg3.ys.astype(float))]
d4 = [0 if x>=thresh else x for x in scipy.fftpack.dct(seg4.ys.astype(float))]
compressed = thinkdsp.Wave(scipy.fftpack.idct(d1),wave.framerate) | thinkdsp.Wave(scipy.fftpack.idct(d2),wave.framerate) | thinkdsp.Wave(scipy.fftpack.idct(d3),wave.framerate) | thinkdsp.Wave(scipy.fftpack.idct(d4),wave.framerate)
compressed.apodize()
compressed.normalize()
```

```python
wave.make_audio()
```

```python
compressed.make_audio()
```

It seems that the algorithm is ok at compressing the audio, even with a super high threshold, it's easy to tell after compression that the audio is still "allen, allen, allen, allen". More segments could probably be a way to show the effects better, but this does give the gist of the idea.
