#### 4) Apply a few different filters to the segment (or the whole wave) and listen to the effect.  Try a low pass, high pass, and band-stop filter.  Try some different cutoff frequencies.

##### Low-pass filter

```python
base_spec = wave.make_spectrum()
lp_spec = base_spec
lp_spec.low_pass(1500)
lpw = lp_spec.make_wave()
lpw.normalize()
lpw.apodize()
lpw.plot()
```

```python
lpw.make_audio()
```

The low-pass filter sounds interesting, as though hearing the goat from through a wall. The high pitches are muffled out and only the bottom parts of the goat's voice can be heard. As the cutoff decreases, you hear less and less of the goats voice, and only thuds.

##### Hi-pass Filter

```python
base_spec = wave.make_spectrum()
hp_spec = base_spec
hp_spec.high_pass(2000)
hpw = hp_spec.make_wave()
hpw.normalize()
hpw.apodize()
hpw.plot()
```

```python
hpw.make_audio()
```

The hi-pass filter makes the screechy and piercing parts of the goat's voice more apparent. Also, certain background scratches and noises are made more audible, as well as an ominous ringing which I was not aware of before.

##### Band-stop Filter

```python
base_spec = wave.make_spectrum()
bs_spec = base_spec
bs_spec.band_stop(500,4000)
bsw = bs_spec.make_wave()
bsw.normalize()
bsw.apodize()
bsw.plot()
```

```python
bsw.make_audio()
```

By looking specifically at the major frequencies which showed up in the spectrum from problem 3, I found that by making my band from 500 to 4k hz, I could effectively almost remove the goat's scream from the audio clip. This is actually pretty neat, considering most of the noise is still there.

#### 5) Synthesize a wave by creating a spectrum with arbitrary harmonics, invoking make_wave, and listening. What happens as you add frequency components that are not multiples of the fundamental?

```python
part1 = thinkdsp.SinSignal(freq=220, amp=1.0, offset=0)
part2 = thinkdsp.SinSignal(freq=440, amp=1.0, offset=0)
comb = part1+part2
wave = comb.make_wave()
wave.make_audio()
```

These two frequencies are octaves apart, and thus they sound the same note-wise. It is hard to pick them apart, only a trained musical ear can tell their are two notes here.

```python
part3 = thinkdsp.SinSignal(freq=880, amp=1.0, offset=0)
comb += part3
wave = comb.make_wave()
wave.make_audio()
```

Adding a higher octave at 880 Hz, it is audible when comparing it with the previous combo, but it's clear that there is no dissonance at play here - the notes sound together and can't be pryed apart by our hearing.

```python
part4 = thinkdsp.SinSignal(freq=350, amp=1.0, offset=0)
comb += part4
wave = comb.make_wave()
wave.make_audio()
```

Add in an awkward 350 hz, and it sounds a bit odd, and also a bit familiar (Can you guess where this is from?). The presence of the 350 stands out and also makes the other frequencies more discernable. 350 and 440 together form the dial tone that landlines use, hence the vague familiarity!

#### 8) Sample an 1100 Hz triangle at 10000 frames per second and listen to it. Can you hear the aliased harmonic? It might help if you play a sequence of notes with increasing pitch.

```python
sawt = thinkdsp.TriangleSignal(freq=1100, amp=1.0, offset=0)
sawtwave = sawt.make_wave(duration=1, start=0, framerate=10000)
sawtwave.make_spectrum().plot()
sawtwave.make_audio()
```

```python
sawt = thinkdsp.TriangleSignal(freq=8900, amp=1.0, offset=0)
sawtwave = sawt.make_wave(duration=1, start=0, framerate=10000)
sawtwave.make_spectrum().plot()
sawtwave.make_audio()
```

The aliasing is apparent as you look at the fundamental's harmonic progress. As you fold back over 5000 hz (half that of the sampling rate, you'll notice that harmonic 5500 folds back to 4500, 6600 to 3400, 7700 to 2300 and so on until you get to 100 which is 9900 folded back. Looking closely, these diminishing spikes in frequencies are apparent on the spectrum. Additionally, sampling an 8900 triangle waves yields the exact same result as the 1100, due to this folding around the 5000. Neat!

#### 10)  The triangle and square waves have odd harmonics only; the sawtooth wave has both even and odd harmonics. The harmonics of the square and sawtooth waves drop off in proportion to 1/f; the harmonics of the triangle wave drop off like 1/f2. Can you find a waveform that has even and odd harmonics that drop off like 1/f2?

```python
part1 = thinkdsp.SinSignal(freq=100, amp=1)
part2 = thinkdsp.SinSignal(freq=200, amp=.25)
part3 = thinkdsp.SinSignal(freq=300, amp=1.0/9)
part4 = thinkdsp.SinSignal(freq=400, amp=1.0/16)
part5 = thinkdsp.SinSignal(freq=500, amp=1.0/25)
part6 = thinkdsp.SinSignal(freq=600, amp=1.0/36)
part7 = thinkdsp.SinSignal(freq=700, amp=1.0/49)
part8 = thinkdsp.SinSignal(freq=800, amp=1.0/64)
mix = part1 + part2 + part3 + part4 + part5 + part6 + part7 + part8
newwave = mix.make_wave(duration=.05)
newspec = newwave.make_spectrum()
newspec.plot()
```

Though I could have created a loop to generate the 1/x^2 drop off for the frequencies and put it into the a Spectrum object, I decided instead to stack various sinusoidal waves over each other because I thought it would be more interesting.
